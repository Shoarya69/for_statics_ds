```python
import pandas as pd
import numpy as np
import scipy.stats as ss
```

```python
df = pd.read_csv('cars_output.csv')
cars = df.copy()
```

```python
cars = cars[
    (cars.yearOfRegistration <=2018) &
    (cars.yearOfRegistration >=1950) &
    (cars.price >= 100) &
    (cars.price <= 150000) &
    (cars.powerPS >= 10) &
    (cars.powerPS <=500)
]

```

```python
sample_size = 1000
sample = cars.sample(sample_size,random_state=0)

sample
```

```python
count = sample['gearbox'].value_counts().get('automatic',0)

nobs = len(sample['gearbox'])

p0 = 0.23
```

```python
print(count,nobs,p0)
```

```python
from statsmodels.stats.proportion import proportions_ztest

stati,pvalue_oneprop =  proportions_ztest(count=count,nobs=nobs,value=p0,alternative='two-sided',prop_var=False)
print(stati,pvalue_oneprop)
```

```python
alpha = 0.05
cv = ss.norm.ppf([alpha/2,1-alpha/2])
print(cv)
```

```python
if pvalue_oneprop <= alpha:
    print("Reject the null hypothesis — proportion has changed.")
else:
    print("Fail to reject the null hypothesis — no significant change.")
```
